```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from traffic_scores import *
```

```python
traffic = load_all_traffic()
kwh = load_kwh()
```

```python
seg_day_scores = traffic.groupby(["Date", "Segment ID"]).agg({"Score": 'mean'})
```

```python
seg_day_scores = seg_day_scores.reset_index()
seg_day_scores.head()
```

```python
seg_day_scores = seg_day_scores.pivot(index="Date", columns="Segment ID", values=["Score"])
```

```python
seg_day_scores
```

```python
seg_day_scores.head()
```

```python
seg_day_scores.columns
```

```python
kwh = kwh[['Date', 'kwh']]
```

```python
merged = pd.merge(kwh, seg_day_scores, on="Date")
```

```python
merged.head()
```

```python
merged = merged.fillna(merged.mean())
```

```python
X = merged.drop(['kwh', 'Date'], axis=1)
y = merged['kwh']
```

```python
X[X.columns[X.isna().any(axis=0)]].columns
```

```python
Xn = X.drop(X[X.columns[X.isna().any(axis=0)]].columns, axis=1)
```

```python
Xn.isna().any(axis=1)
```

```python
from sklearn.linear_model import Lasso, LassoCV, LassoLars
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Xn, y, test_size=0.3, random_state=101)
lm = LassoLars()
lm.fit(X_train, y_train)

print(lm.coef_)

print(lm.intercept_)
```

```python
from sklearn.feature_selection import SelectKBest

model = SelectKBest()
X_new = model.fit_transform(Xn, y)
model.get_support()
```

```python
segments = []
for i in X_new.columns:
	segments.append(i[1])
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold

model = XGBRegressor(n_estimators=1000, max_depth=7, eta=0.1, subsample=0.7, colsample_bytree=0.8)
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate model
scores = cross_val_score(model, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
```

```python
from numpy import absolute

scores = absolute(scores)
print('Mean MAE: %.3f (%.3f)' % (scores.mean(), scores.std()) )
```

```python
model.fit(X_train, y_train)

preds1 = model.predict(X_test)

plt.scatter(preds1, y_test)
```

```python
from scipy.stats import linregress

x = preds1
y=y_test
gradient, intercept, r_value, p_value, std_err = linregress(preds1, y_test)
mn=np.min(x)
mx=np.max(x)
x1=np.linspace(mn,mx,500)
y1=gradient*x1+intercept
plt.plot(x,y,'ob')
plt.plot(x1,y1,'-r')
plt.ylabel("KWH")
plt.xlabel("Score")
plt.show()
```
