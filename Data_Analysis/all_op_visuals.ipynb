```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read in the CSV file
alloperators = pd.read_csv("./block_weather_avg_test.csv")
```

```python
# show the most common operator types
print("Operator types most often used")
alloperators['op_type'].value_counts().head(5)
```

```python
#Plot the Avg MPP vs Block
sns.set(rc = {'figure.figsize':(15,4)})
sns.barplot(data=alloperators, x="block", y="mpp", color='violet', order=alloperators.groupby('block')['mpp'].agg('mean').sort_values(ascending=True).index)
```

```python
#Plot the KWH vs Block
sns.barplot(data=alloperators, x="block", y="kwh", color='violet', order=alloperators.groupby('block')['kwh'].agg('mean').sort_values(ascending=False).index)
```

```python
#block vs. visibility miles
sns.barplot(data=alloperators, x="block", y="visibilityMiles", color='violet', order=alloperators.groupby('block')['visibilityMiles'].agg('mean').sort_values(ascending=True).index)
```

```python
# Plot the block vs. tempF
sns.barplot(data=alloperators, x="block", y="tempF", color='violet', order=alloperators.groupby('block')['tempF'].agg('mean').sort_values(ascending=True).index)
```

```python
# Plot the block vs. tempF
sns.barplot(data=alloperators, x="block", y="pressureInches", color='violet', order=alloperators.groupby('block')['pressureInches'].agg('mean').sort_values(ascending=True).index)
```
