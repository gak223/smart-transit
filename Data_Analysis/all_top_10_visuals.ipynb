```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read in the CSV file
df = pd.read_csv("./block_weather_avg_test.csv")
```

```python
print("Operator IDs most often used")
df['op_id'].value_counts().head(10)
```

```python
#Plot the Avg MPP vs Block
sns.set(rc = {'figure.figsize':(15,4)})
sns.barplot(data=df, x="block", y="mpp", color='violet', order=df.groupby('block')['mpp'].agg('mean').sort_values(ascending=True).index)
```

```python
#Plot the KWH vs Block
sns.barplot(data=df, x="block", y="kwh", color='violet', order=df.groupby('block')['kwh'].agg('mean').sort_values(ascending=False).index)
```

```python
#Operator vs. TAVG
sns.barplot(data=df, x="op_id", y="TAVG", color='violet', order=df.groupby('op_id')['TAVG'].agg('mean').sort_values(ascending=True).index)
```

```python
# Plot the operator ID vs. TMAX
sns.set(rc = {'figure.figsize':(10,4)})
sns.barplot(data=df, x="op_id", y="TMAX", color='violet', order=df.groupby('op_id')['TMAX'].agg('mean').sort_values(ascending=True).index)
```

```python
# Plot the operator ID vs. TMIN
sns.set(rc = {'figure.figsize':(10,4)})
sns.barplot(data=df, x="op_id", y="TMIN", color='violet', order=df.groupby('op_id')['TMIN'].agg('mean').sort_values(ascending=True).index)
```
